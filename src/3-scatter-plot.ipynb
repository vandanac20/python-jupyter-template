### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt

# Data
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# Error Handling
if not isinstance(x_scatter, list) or not isinstance(y_scatter, list):
    raise TypeError("x_scatter and y_scatter must be lists")
if len(x_scatter) == 0 or len(y_scatter) == 0:
    raise ValueError("x_scatter and y_scatter cannot be empty")
if len(x_scatter) != len(y_scatter):
    raise ValueError("x_scatter and y_scatter must have the same length")

# Create the scatter plot
plt.scatter(x_scatter, y_scatter)

# Annotate each point with its (x, y) value
for i in range(len(x_scatter)):
    plt.annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(0, 5), ha='center')

# Set title
plt.title('Scatter Plot with Annotations')

# Labeling the axes
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.show()
```
